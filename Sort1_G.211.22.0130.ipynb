```python
# Bubble Sort
def bubbleSort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Driver code to test bubbleSort
arr_bubble = [10, 30, 20, 40, 50, 70, 60]
bubbleSort(arr_bubble)
print("Sorted array (Bubble Sort):", arr_bubble)

# Selection Sort
def selectionSort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# Driver code to test selectionSort
arr_selection = [1, 45, 32, 62, 71]
selectionSort(arr_selection)
print("Sorted array (Selection Sort):", arr_selection)

# Insertion Sort
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Driver code to test insertionSort
arr_insertion = [12, 11, 13, 5, 6]
insertionSort(arr_insertion)
print("Sorted array (Insertion Sort):", arr_insertion)
```
